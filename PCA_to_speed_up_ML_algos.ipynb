## PCA to Speed Up Machine Learning Algorithms

- Objectives:
    - Determine an optimal number of components for PCA
    - Apply PCA to a supervised machine learning model
    - Compare the speed of a model trained with the entire feature set vs. principal components

### Imports, Loading, Custom Functions

**Imports**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline

pd.set_option('display.max_columns',200)
pd.set_option("display.max_info_rows", 800)
pd.set_option('display.max_info_columns',800)

from sklearn import set_config
set_config(transform_output='pandas')
```

**Custom Functions**

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

def classification_metrics(y_true, y_pred, label='',
                           output_dict=False, figsize=(8,4),
                           normalize='true', cmap='Blues',
                           colorbar=False):
    # Get the classification report
    report = classification_report(y_true, y_pred)
    ## Print header and report
    header = "-"*70
    print(header, f" Classification Metrics: {label}", header, sep='\n')
    print(report)
    ## CONFUSION MATRICES SUBPLOTS
    fig, axes = plt.subplots(ncols=2, figsize=figsize)
    # create a confusion matrix  of raw counts
    ConfusionMatrixDisplay.from_predictions(y_true, y_pred,
                normalize=None, cmap='gist_gray', colorbar=colorbar,
                ax = axes[0],);
    axes[0].set_title("Raw Counts")
    # create a confusion matrix with the test data
    ConfusionMatrixDisplay.from_predictions(y_true, y_pred,
                normalize=normalize, cmap=cmap, colorbar=colorbar,
                ax = axes[1]);
    axes[1].set_title("Normalized Confusion Matrix")
    # Adjust layout and show figure
    fig.tight_layout()
    plt.show()
    # Return dictionary of classification_report
    if output_dict==True:
        report_dict = classification_report(y_true, y_pred, output_dict=True)
        return report_dict


def evaluate_classification(model, X_train, y_train, X_test, y_test,
                         figsize=(6,4), normalize='true', output_dict = False,
                            cmap_train='Blues', cmap_test="Reds",colorbar=False):
    # Get predictions for training data
    y_train_pred = model.predict(X_train)
    # Call the helper function to obtain regression metrics for training data
    results_train = classification_metrics(y_train, y_train_pred, #verbose = verbose,
                                     output_dict=True, figsize=figsize,
                                         colorbar=colorbar, cmap=cmap_train,
                                     label='Training Data')
    print()
    # Get predictions for test data
    y_test_pred = model.predict(X_test)
    # Call the helper function to obtain regression metrics for test data
    results_test = classification_metrics(y_test, y_test_pred, #verbose = verbose,
                                  output_dict=True,figsize=figsize,
                                         colorbar=colorbar, cmap=cmap_test,
                                    label='Test Data' )
    if output_dict == True:
        # Store results in a dataframe if ouput_frame is True
        results_dict = {'train':results_train,
                    'test': results_test}
        return results_dict
```

**Loading**

```python
# Get Data
df = pd.read_csv('Data/pd_speech_features.csv')
df.head()
```

### Prepare Dataset for Machine Learning

```python
df.dtypes.value_counts()
```

- All the data is numeric; check for null values.

```python
df.isna().sum().sum()
```

**Define X and y**

```python
## Define the target and drop the old id column
# Define target 
target_col = 'class'
drop_cols = ['id']
# Define X and y
y = df[target_col].copy()
X = df.drop(columns=[target_col, *drop_cols]).copy()
```

**Train Test Split**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=321)
X_train.head()
```

**Scale Data**
- Data should always be scaled before applying PCA. PCA is a distance-based algo

```python
# Instantiate StandardScaler
scaler = StandardScaler()
# Fit and transform data
X_train_tf = scaler.fit_transform(X_train)
X_test_tf = scaler.transform(X_test)
```

**Check for Class Balance**

```python
# Check for class balance in target
y_train.value_counts(1)
```

- There are unbalanced classes, which will be addressed with SMOTE

```python
from imblearn.over_sampling import SMOTE, SMOTENC
smote = SMOTE()
X_train_sm, y_train_sm = smote.fit_resample(X_train_tf, y_train)
y_train_sm.value_counts()
```

### Time a Model Without PCA

- Because the goal is to demonstrate how PCA can speed up a model's training, a way to measure it is required.
    - Import datetime library and use dt.datetime.now() to record the start and end times, then subtract to find the duration.

```python
# Import datetime library
import datetime as dt

# Record the start time
start = dt.datetime.now()
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train_sm, y_train_sm)

# Record the end time and calc duration
end = dt.datetime.now()
dur_baseline = end-start

evaluate_classification(clf, X_train_sm, y_train_sm, X_test_tf, y_test)
print(f'Training time was: {dur_baseline}')
```

- This gives us a baseline time for how long it takes to train the model using all of the original features.

### Time a Model Using PCA

- By default, PCA will return the same number of PCs as the number of original features. Together, these PCs will explain 100% of the variance.
- However, this doesn't serve the purpose of dimensionality reduction, and won't speed up any modeling.
- By dropping the number of features/PCs used, some data will be lost for the sake of speed.
- The goal is to find the best balance of sacrificing info for speed.
- For the example below, the number of components will be manually set to three.

```python
# Instantiate and fir data using PCA
pca = PCA(n_components=3)
X_train_pca = pca.fit_transform(X_train_sm)
X_test_pca = pca.transform(X_test_tf)
X_train_pca.head()
```

**Time a model with 3 PCs**

```python
# Record the state time
start = dt.datetime.now()

clf = RandomForestClassifier(random_state=42)
clf.fit(X_train_pca, y_train_sm)

# Record the end time and calc duration
end = dt.datetime.now()
dur_pca = end-start

evaluate_classification(clf, X_train_pca, y_train_sm, X_test_pca, y_test)
print(f'Training time was: {dur_pca}')
```

- Finally, compare the speed of the two tests.

```python
# Compare the speeds before and after PCA
compare_speed = dur_baseline/dur_pca
print(f'Using PCs was {compare_speed:.2f} times faster!')
```

- This model is significantly faster, however, it did not perform as well.

### Choosing the Number of Components to Return

- Choosing more components will always increase the ammount of variance explained by the model.
- Cumulative variance can be visualized with the following code.

```python
# First, initialize and fit PCA
pca = PCA()
pca.fit(X_train_sm, y_train)
```

```python
# Next obtain the ratio of variance explained by each PC
explained = pd.Series(pca.explained_variance_ratio_, name='Explained Variance Ratio')
explained
```

- To better understand, visualize the amount of variance explained as the number of PCs increases.
- To do this, plot the cumulative sum. For reference a line at 90% will be included.

```python
# Plot the cumulative sum of the percentage of explained variance
ax = explained.cumsum().plot(marker='.')
# Add a line to mark 90%
ax.axhline(0.9, color='k')
```

- Initially the amount of variance explained by PCs rises quite dramatically, but the returns start to diminish.

**Specifying the Explained Variance**
- Rather than choose a number of components, another option is to specify the minimum proportion of explained variance that is acceptabel.
- PCA will automatically reduce the number of PCs just enough to meet the requirements.
- To specify a proportion, give n_components a float between 0 and 1 for the percentage.

```python
# Define PCA to address 85% of variance
pca85 = PCA(n_components=0.85)
# Fit and transform on training data
X_train_pca85 = pca85.fit_transform (X_train_sm)
# Transform test data
X_test_pca85 = pca85.transform(X_test_tf)
# Obtain the number of PCs used
pca85.n_components_
```

**Timing when 85% of variance is explained with PCs**

```python
# Record the start time
start = dt.datetime.now()

clf = RandomForestClassifier(random_state=42)
clf.fit(X_train_pca85, y_train_sm)

# Record the end time and calc duration
end = dt.datetime.now()
dur_pca_85 = end - start

evaluate_classification(clf, X_train_pca85, y_train_sm, X_test_pca85, y_test)
print(f'Training time was: {dur_pca_85}')
```

```python
# Compare the speed of the model using PCA explaining 85% of the model vs the model using all features
compare_speed = dur_baseline/dur_pca_85
print(f"Using PC's with 0.85 was {compare_speed:.2f} times faster!")
```

### Summary
- While PCA is an unsupervised modeling technique, it can be used to prepare data for predictive modeling by reducing the number of features.
- PCA can be used in a pipeline along other transformers and models. It can speed up training and will sometimes improve predictions.
- Data should always be sdcaled before applying PCA or an other distance-based models.
